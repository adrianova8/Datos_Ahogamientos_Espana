## <span style=color:darkblue> Descomposición serie temporal </span>

## <span style=color:darkblue>  Series temporales estacionarias/estacionarias </span>

+ La estacionariedad es una propiedad de una serie de tiempo. Una serie estacionaria es aquella en la que los valores de la serie no son una función del tiempo.

+ La mayoría de los métodos de pronóstico estadístico están diseñados para trabajar en una serie de tiempo estacionaria. El primer paso en el proceso de pronóstico suele ser hacer alguna transformación para convertir una serie no estacionaria en estacionaria.

### <span style=color:darkblue>  Probar la estacionalidad </span>

+ Usaremos la prueba ADF (Aumented Dickey-Fuller), donde la hipótesis nula es que la serie de tiempo posee raíz unitaria y no es estacionaria. Entonces, si el p-valor en la prueba es menor que el nivel de significación (0.05), rechaza la hipótesis nula.


+ Usaremos también la prueba KPSS que se utiliza para probar la estacionariedad de la tendencia. La hipótesis nula y la interpretación del valor P es justo lo contrario de la prueba Dickey-Fuller. 

```python
# Prueba Dickey-Fuller
result = adfuller(ahogamientos_desc.Ahogamientos.values, autolag = 'AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

+ **ADF Statistic:** El valor de -4.669744720883896 es la estadística del test ADF. Si este valor es más negativo que los valores críticos, proporciona evidencia en contra de la hipótesis nula de que la serie temporal no es estacionaria. En este caso, el valor de -4.669744720883896 indica que la serie temporal es probablemente estacionaria.


+ **p-value ADF:** El valor de p, 9.588341394894362e-05, es el p-valor asociado al estadístico ADF. Representa la probabilidad de obtener un estadístico ADF tan extremo como el observado si la hipótesis nula fuera verdadera. Un valor de p bajo (generalmente por debajo de un umbral de significancia, como 0.05) indica que la serie temporal es estacionaria. En este caso, el valor extremadamente bajo del p-valor (9.588341394894362e-05) sugiere que la serie temporal es estacionaria.


## <span style=color:darkblue>  Modelo SARIMA </span>

+ SARIMA (Seasonal Autoregressive Integrated Moving Average), es una clase de modelo que explica una serie de tiempo determinada en función de los valores pasados, es decir, sus proporios retrasos y errores de pronóstico retrasados, por lo que se puede usar para pronosticar valores futuros. 


+ Nuestro modelo presenta estacionalidad, por lo tanto debemos aplicar SARIMA en vez de ARIMA, ya que este último no contempla la estacionalidad. 


+ Un modelo SARIMA se compone de 3 términos:

    + **p** es el orden del término AR

    + **q** es el orden del término MA

    + **d** es el número de diferenciaciones requeridas para hacer que la serie de tiempo sea estacionaria

### <span style=color:darkblue>  Diferenciación (d) en el modelo SARIMA </span>

+ El orden correcto de diferenciación es la diferenciación mínima requerida para obtener una serie casi estacionaria que se desplaza alrededor de una media definida y la gráfica ACF llega a cero bastante rápido.


+ Si las autocorrelaciones son positivas para muchos rezagos (10 o más), entonces la serie necesita más diferenciación. Por otro lado, si la autocorrelación del retardo 1 en sí misma es demasiado negativa, es probable que la serie esté sobrediferenciada.

```python
# Gráfico de autocorrelación 
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':120})
fig, axes = plt.subplots(2, 2)

# Ajustar el espaciado vertical entre subplots
plt.subplots_adjust(hspace=0.5)

# Original Series
axes[0, 0].plot(ahogamientos_desc.Ahogamientos); axes[0, 0].set_title('Original Series')
plot_acf(ahogamientos_desc.Ahogamientos.dropna(), ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(ahogamientos_desc.Ahogamientos.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(ahogamientos_desc.Ahogamientos.diff().dropna(), ax=axes[1, 1])

plt.show()
```

```python
# Numero de diferenciaciones según diferentes test
y = ahogamientos_desc.Ahogamientos

## Adf Test
print('ADF Test: ', ndiffs(y, test='adf'))  
```

+ Podemos observar como a partir del Test de Dickey Fuller realizado anteriormente nos daba que los datos muestran estacionariedad, por lo tanto no es necesario aplicar diferenciación. Vemos como los datos originales tienden rápido a 0 cuando dibujamos el plot de autocorrelaciones sin aplicar órdenes de diferenciación, mientras que si aplicamos una primera diferenciación vemos como el retraso pasa bastante rápido a la zona negativa lejana, lo que indica que la serie podría haber sido sobrediferenciada.

- Nos quedaremos con diferencición d=0.

### <span style=color:darkblue>  Autoregresivo (AR) en el modelo SARIMA </span>

+ Un modelo autorregresivo puro (solo AR) es aquel en el que Yt depende solo de sus propios retrasos. Es decir, Yt es una función de los 'retrasos de Yt'. 


+ Puede averiguar el número necesario de términos AR inspeccionando el gráfico de autocorrelación parcial (PACF). PACF transmite la correlación pura entre un retraso y la serie. De esa manera, sabrá si ese retraso es necesario en el término AR o no.

```python
# Encontrar el orden del término AR (p)

# Gráfico de autocorrelación 
plt.rcParams.update({'figure.figsize':(10,3), 'figure.dpi':120})
fig, axes = plt.subplots(1, 2)

# Original Series
axes[0].plot(ahogamientos_desc.Ahogamientos); axes[0].set_title('Original Series')
plot_pacf(ahogamientos_desc.Ahogamientos.dropna(), ax=axes[1])

plt.show()
```

+ Podemos observar como el PACF lag 1 es bastante significativo, ya que está muy encima de la línea de significación. El lag 2 también resultado significativo al igual que el lag 3 y el lag 4, logrando cruzar ligeramente el límite de significación (región azul). Voy a coger hasta 2 retrasos para ser precavidos.

### <span style=color:darkblue>  promedio móvil puro (MA) en el modelo SARIMA </span>

+ Un modelo de promedio móvil puro (solo MA) es uno en el que Yt depende solo de los errores de pronóstico retrasados 


+ El ACF indica cuántos términos MA se requieren para eliminar cualquier autocorrelación en la serie estacionaria.

```python
# Encontrar el orden del término MA (q)

# Gráfico de autocorrelación 
plt.rcParams.update({'figure.figsize':(10,3), 'figure.dpi':120})
fig, axes = plt.subplots(1, 2)

# Original Series
axes[0].plot(ahogamientos_desc.Ahogamientos.values); axes[0].set_title('Original Series')
plot_acf(ahogamientos_desc.Ahogamientos.dropna(), ax=axes[1])

plt.show()
```

+ El primer retraso está muy por encima de la línea de significación, mientras que el resto de retrasos se mantienen parcialmente fuera no tienen tanta significancia. Por lo tanto fijaremos q = 1. Ante la duda es mejor elegir el modelo más simple que exlique suficientemente la Y.

### <span style=color:darkblue>  Construcción del modelo </span>

+ El modelo se representará como SARIMA(p,d,q)x(P,D,Q), donde, P, D y Q son términos SAR, orden de diferenciación estacional y SMA respectivamente y es la frecuencia del 'x'tiempo serie.


+ Si su modelo tiene patrones estacionales bien definidos, aplique D=1 para una frecuencia dada 'x'.


+ Como regla general, configure los parámetros del modelo de modo que D nunca exceda uno. Y la diferenciación total 'd + D' nunca excede 2. Trate de mantener solo los términos SAR o SMA si su modelo tiene componentes estacionales.

```python
# Plot
fig, axes = plt.subplots(2, 1, figsize=(10,5), dpi=100, sharex=True)

# Usual Differencing
axes[0].plot(ahogamientos_desc[:], label='Original Series')
axes[0].plot(ahogamientos_desc[:].diff(1), label='Usual Differencing')
axes[0].set_title('Usual Differencing')
axes[0].legend(loc='upper left', fontsize=10)


# Seasinal Dei
axes[1].plot(ahogamientos_desc[:], label='Original Series')
axes[1].plot(ahogamientos_desc[:].diff(12), label='Seasonal Differencing', color='green')
axes[1].set_title('Seasonal Differencing')
plt.legend(loc='upper left', fontsize=10)
plt.show()
```

+ Es invariable que apliquemos diferenciación habitual (log 1) que diferenciación estacional (log 12), ya que los datos son estacionarios y por lo tanto, media y varianza son constantes a lo largo del tiempo, excepto por el hecho que se observa claramente la estacionalidad, debido al gran número de ahogamientos que se produce durante los meses de verano.

```python
# Seasonal - fit stepwise auto-ARIMA
smodel = pm.auto_arima(ahogamientos_desc.Ahogamientos, start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3, m=12,
                         start_P=0, seasonal=True,
                         d=1, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

smodel.summary()
```

+ Como los datos abarcan todos los meses del año (enero a diciembre) y la estacionalidad se presenta en los meses de junio a septiembre, entonces el valor de 'm' será 12. Esto se debe a que el patrón estacional se repite cada 12 meses, correspondiendo a los meses de verano.

+ El mejor modelo tiene un AIC de 11163.238 y los valores de P son significativos.SARIMAX(1, 0, 1)x(2, 1, 0, 12)

```python
# Gráficos de los residuos en el modelo SARIMA
smodel.plot_diagnostics(figsize= (10,8))
plt.show()
```

**Plot1:** los errores residuales parecen fluctuar alrededor de una media de cero y tienen una varianza uniforme.

**Plot2:** el gráfico de densidad sugiere una distribución normal con media cero.

**Plot3:** Todos los puntos deben estar perfectamente alineados con la línea roja. Excepto por la última parte de los puntos que aparece una desviación significativa que implica que la distribución puede estar sesgada.

**Plot4:** Abajo a la derecha: el gráfico Correlogram, también conocido como ACF, muestra que los errores residuales no están autocorrelacionados. Cualquier autocorrelación implicaría que existe algún patrón en los errores residuales que no se explican en el modelo. Por lo tanto, deberá buscar más X (predictores) para el modelo.
