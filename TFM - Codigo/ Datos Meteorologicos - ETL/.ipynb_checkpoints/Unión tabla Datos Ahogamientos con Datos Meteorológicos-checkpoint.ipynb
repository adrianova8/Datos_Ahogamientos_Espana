# Unión tabla Datos Ahogamientos con Datos Meteorológicos

```python
# Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Datos ahogamientos

```python
# Datos ahogamiento
datos_ahogamiento = pd.read_csv("datos-ahogamiento.csv")

# Limpiamos las variables vacías
datos_ahogamiento = datos_ahogamiento.iloc[:, :27] # quitamos las columnas vacías de información
```

```python
# Forma
datos_ahogamiento.shape
```

```python
# Primeras filas
datos_ahogamiento.head()
```

```python
# Tipos de datos
datos_ahogamiento.dtypes
```

# Datos Meteorologicos

```python
# Datos meteorologicos
datos_meteorologicos = pd.read_csv("datos_meteorologicos.csv")
```

```python
datos_meteorologicos.shape
```

```python
datos_meteorologicos
```

```python
datos_meteorologicos.dtypes
```

```python
# Buscando diferencias entre columnas Fecha y Provincia

# Fecha
ahog = set(datos_ahogamiento["Fecha"])
met = set(datos_meteorologicos["Fecha"])

ahog.symmetric_difference(met)

# Provincia
#ahog = set(datos_ahogamiento["Provincia"])
#met = set(datos_meteorologicos["Provincia"])

#ahog.symmetric_difference(met)
```

```python
# Hay que convertir la variable fecha para que sea igual en ambos dataframes para que al hacer el match coincida correctamente
datos_ahogamiento["Fecha"] = pd.to_datetime(datos_ahogamiento["Fecha"], utc = True)
datos_meteorologicos["Fecha"] = pd.to_datetime(datos_meteorologicos["Fecha"], utc = True)

# Extraer el día, mes y año en un nuevo formato de fecha
datos_ahogamiento["Fecha"] = datos_ahogamiento["Fecha"].dt.strftime("%d-%m-%Y")
datos_meteorologicos["Fecha"] = datos_meteorologicos["Fecha"].dt.strftime("%d-%m-%Y")
```

## Creamos el JOIN

```python
# Dataframe Final 
data = pd.merge(datos_ahogamiento, datos_meteorologicos, on=['Fecha', 'Provincia'], how='inner')
data.shape
```

```python
# Eliminamos la última coluna que no aporta información
data = data.iloc[:, :-1]
```

```python
# Guardamos el dataset final
data.to_csv('datos_ahogamiento_meteorologicos.csv', index=False)
```

# ¿POR QUÉ TENGO MÁS DATOS EN LA BASE DE DATOS FINAL QUE EN LA BASE DE DATOS DE AHOGAMIENTOS?

```python
# Forma de las 3 bases de datos
print("Datos de Ahogamiento:", datos_ahogamiento.shape)
print("Datos Meteorológicos:", datos_meteorologicos.shape)
print("Datos Final:", data.shape)
```

+ Las 27 + 16 = 43 columnas que deberían aparecer en la base de datos final, solo aparecen 40, ya que Fecha y Provincia son las variables que se emparejan entre las bases de datos y solo aparece una vez. Además aparece una columna extra que es el recuento de todos los datos que hay para una estacion, que no aporta información y nos deshacemos de ella.

```python
# Duplicados datos_meteorologicos 
datos_meteorologicos[datos_meteorologicos.duplicated(subset=['Fecha', 'Provincia'])]
```

Parece que la base de datos meteorologicos tiene duplicados en [Fecha y Provincia] para 8766 registros, esto puede hacer que en el momento que un registro de ahogamiento se empareje con un dato meteorologico, se acabe duplicando x2, de ahí que tengamos más observaciones la base de datos final.

DATOS METEOROLOGICOS:

   - duplicados (Fecha y Provincia): 8766 registros
   - base de datos meteorologicos: 160674 registros

```python
# Creamos una copia de la base de datos meteorologicos
datos_meteorologicos_copia = datos_meteorologicos.copy()
datos_meteorologicos_copia
```

```python
# Eliminamos los duplicados. Ahora tenemmos 151908 registros para datos_ahogamientos
datos_meteorologicos_copia = datos_meteorologicos_copia.drop_duplicates(subset=['Fecha', 'Provincia'], keep='first')
```

```python
# Probamos de volver a hacer el JOIN
# Dataframe Final 
data2 = pd.merge(datos_ahogamiento, datos_meteorologicos_copia, on=['Fecha', 'Provincia'], how='inner')
data2.shape
```

```python
# Eliminamos la última coluna que no aporta información
data2 = data2.iloc[:, :-1]
```

```python
# Guardamos el dataset final
data2.to_csv('datos_ahogamiento_meteorologicos2.csv', index=False)
```
